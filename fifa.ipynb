```python
def height_into_cm(height_str):
  try:
    feet, inches = height_str.split("'")
    feet = int(feet.strip())
    inches = int(inches.strip().replace('"',''))
    return feet*30.48 + inches* 2.54
  except:
    return None
df['height_cm'] = df['height'].apply(height_into_cm)
```

```python
def hits_k_to_num(x):
  if isinstance(x,str):
    x = x.lower().strip()
    if 'k' in x:
      return float(x.replace('k',""))*1000
    else:
      return pd.to_numeric(x, errors='coerce')
  return x

df['hits'] = df['hits'].apply(hits_k_to_num)
```

```python
import re
def value_cleanup(v):
  if isinstance(v,str):
    v= v.lower().strip()
    if 'm' in v:
      v = re.sub(r'[€m]','',v)
      return float(v)*1000000
    elif 'k' in v:
      v = re.sub(r'[€k]','',v)
      return float(v)*1000
    else:
      v= v.replace('€','')
      return float(v)
  return v
df['value'] = df['value'].apply(value_cleanup)
```
